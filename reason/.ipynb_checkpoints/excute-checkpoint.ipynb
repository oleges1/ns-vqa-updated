```python
def make_vocab(programs):
    vocab = defaultdict(dict)
    vocab['question_token_to_idx']['<NULL>'] = 0
    vocab['question_token_to_idx']['<START>'] = 1
    vocab['question_token_to_idx']['<END>'] = 2
    vocab['program_token_to_idx']['<NULL>'] = 0
    vocab['program_token_to_idx']['<START>'] = 1
    vocab['program_token_to_idx']['<END>'] = 2
    vocab['answer_token_to_idx'] = {}
    items = set()
    for program in programs:
        for command in program:
            if not command in vocab['program_token_to_idx']:
                items.add(command)
    for i, item in enumerate(items):
    #     vocab['program_idx_to_token'][i] = item
        vocab['program_token_to_idx'][item] = i + 3
    return vocab

def encode_program(program, vocab):
    return [vocab['program_token_to_idx'][command] for command in program[::-1] + ['<END>']]

def save_vocab(vocab):
    with open(vocab_json, "w") as fp:
        json.dump(vocab , fp) 
```

```python
programs = [
    [
        # шар который находится близко к красному цилиндру 
        'scene',
        'filter_shape[cylinder]',
        'filter_color[red]',
        'unique',
        'relate[close]',
        'filter_shape[sphere]',
        'unique'
    ],
    [
        # шар который находится далеко от цианогого куба 
        'scene',
        'filter_shape[cube]',
        'filter_color[cyan]',
        'unique',
        'relate[far]',
        'filter_shape[sphere]',
        'unique'
    ]
]
```

```python
vocab = make_vocab(programs)
save_vocab(vocab)
```

```python
executor = ClevrExecutor(train_scene_json, val_scene_json, vocab_json)
```

```python
executor.run(encode_program(programs[0], vocab), 3501, 'val', debug=True)
```
